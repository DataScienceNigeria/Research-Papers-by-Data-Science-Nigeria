```python
import folium

def plot_map(df, lat_col='lat', lon_col='lng', popup_col='place'):
  # calculate the bounds of the locations
  bounds = [
    [df[lat_col].min(), df[lon_col].min()],
    [df[lat_col].max(), df[lon_col].max()]
  ]

  # calculate the center of the locations
  center = [
    (df[lat_col].min() + df[lat_col].max()) / 2,
    (df[lon_col].min() + df[lon_col].max()) / 2
  ]

  # create map centered on the center of the locations
  my_map = folium.Map(location=center, zoom_start=11)

  # add a popup for each location in the dataframe
  for i, row in df.iterrows():
    location = [row[lat_col], row[lon_col]]
    # popup = folium.Popup(row[popup_col])
    popup_text = f"Location: {row['place']}<br>\
                    Traffic State: {row['traffic_class']}<br>"
    folium.Marker(location=location, popup=popup_text).add_to(my_map)

  return my_map
```

```python
import collections
import gmaps

def gmap_display(data):

  # Define the center of the map
  center = (data['lat'].mean(), data['lng'].mean())
  locations=data[['lat', 'lng']]

  # Define the figure layout
  fig_layout = {
      'width': '800px',
      'height': '600px',
      'border': '1px solid black',
      'padding': '1px'
  }

  # Create the map figure
  fig = gmaps.figure(
      center=center, 
      zoom_level=12, 
      layout=fig_layout,
      map_type='HYBRID',
  )
  # coord = list(zip(data['lat'], data['lng']))
  # symbols = gmaps.symbol_layer(
  #         coord, fill_color='red', stroke_color='red')
  
  # fig.add_layer(symbols)
  fig.add_layer(gmaps.heatmap_layer(locations))


  # Display the map
  return fig
```
