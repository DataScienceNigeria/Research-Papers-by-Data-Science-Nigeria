```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm


def show_traffic_levels(df, traffic_class_category, traffic_class_column="traffic_class", place_column = "place",  date="date"):
  df = df[df[traffic_class_column] == traffic_class_category]

  date_interval = f'{df[date].min()} - {df[date].max()}'

  # create a list of all the places with traffic_class_category
  places_list = []
  for places in df[place_column]:
      places_list.append(places)

  # count the frequency of each place and create a dataframe
  places_freq = pd.DataFrame(places_list, columns=[place_column])
  places_freq = places_freq.groupby(place_column).size().reset_index(name="count")
  # places_freq = df[place_column].value_counts().reset_index(name="count")
  # places_freq.rename(columns={'index': place_column}, inplace=True)


  # sort the dataframe by frequency in descending order
  places_freq = places_freq.sort_values("count", ascending=False)

  # limit the number of places displayed to the top 20 most frequent places
  places_freq = places_freq.head(20)

  cmap = cm.get_cmap('viridis')

  # plot the bar chart
  sns.set_style("whitegrid")
  fig, ax = plt.subplots(figsize=(12, 8))
  sns.barplot(x="count", y="place", data=places_freq, palette=cmap(places_freq['count']/places_freq['count'].max()))
  ax.set_xlabel("Frequency", fontsize=14)
  ax.set_ylabel("Place", fontsize=14)
  ax.set_title(f"Top 20 Places with ({traffic_class_category}) ({date_interval})", fontsize=16)
  ax.grid(axis='x', linestyle='--', alpha=0.7)

  plt.xticks(rotation=45, ha='right', fontsize=12)
  sns.despine(right=True, top=True)
  ax.legend(title="Frequency", loc='lower right', fontsize=12)
  plt.show()
```
